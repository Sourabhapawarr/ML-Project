```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
sales = pd.read_csv('BigMartSalesData.csv')
sales
```

```python
#1.Plot Total Sales Per Month for Year 2011. How the total sales have increased over months in Year 2011. 
#Which month has lowest Sales?
```

```python
total_month_sale = sales.groupby('Month').agg({'Amount':np.sum})
total_month_sale
```

```python
plt.plot(total_month_sale.index,total_month_sale.values)
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("2011 Sales Report")
plt.show()
```

```python
#2.Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to visualize than Simple Plot?
```

```python
sales.groupby('Month').agg({'Amount':np.sum}).plot(kind='bar', color='orange')
plt.xlabel('Year of 2011')
plt.ylabel('Total Sales')
plt.title('Sales Per Month in 2011')
plt.show()
```

```python
#3. Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?
```

```python
sales_country = sales.groupby('Country').agg({'Amount':np.sum})
sales_country
```

```python
df_grouped_country = sales.query("Year == 2011").filter(
    ["Country", "Amount"]).groupby(["Country"], as_index=False).sum()

plt.pie(df_grouped_country["Amount"],
        labels=df_grouped_country["Country"],
        autopct='%1.1f%%',
        shadow=True)

plt.tight_layout()
plt.show()
```

```python
#4.Plot Scatter Plot for the invoice amounts and see the concentration of amount. 
#In which range most of the invoice amounts are concentrated
```

```python
sales_invoice = sales.groupby('InvoiceNo').agg({'Amount':np.sum})
print (sales_invoice)
plt.scatter(sales_invoice.values,sales_invoice.values,color='b')
plt.grid(True)
plt.show()
```
